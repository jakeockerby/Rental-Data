```python
#Loading necessary packages
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

#Loading the dataframe
rentals = pd.read_excel('/home/jake_ockerby/Python/conrad_data/Flat renting data jake.xlsx')
rentals
```

```python
#Filling in the missing values
rentals = rentals.fillna('Missing')
rentals
```

```python
#Dropping all sources that aren't Airbnb
not_airbnb = rentals[rentals['Source']!='Airbnb'].index
rentals = rentals.drop(not_airbnb)
rentals
```

```python
#Dropping rows where accommodation fee is missing
missing_accom_fee = rentals[rentals['Accommodation Fee']=='Missing'].index
rentals = rentals.drop(missing_accom_fee)
rentals
```

```python
#Dropping rows where listing name is equal to 'Middle'
middle = rentals[rentals['Listing Name']=='Middle'].index
rentals = rentals.drop(middle)
rentals
```

```python
#Dropping empty columns
rentals = rentals.drop(columns=['C-in Time', 'C-out Time', 'Column1'])
rentals
```

```python
# Defining a function which splits guest location into 2, then returns the last segment
def countries(x):
    x = x.split(', ')
    return x[-1]
#Creating a new column which maps the fuction above to obtain a column of just states/countries
rentals['State/Country'] = rentals['Guest Location'].map(countries)
rentals
```

```python
#Defining a function that splits the guest location into 2 and returns the first segment,
#providing that the guest location contains a comma

#The if statement was necessary to distinguish from countries that had a missing city,
#otherwise some of these countries would have been listed as cities
def cities(x):
    if ',' in x:
        x = x.split(', ')
        return x[0]

rentals['Town/City'] = rentals['Guest Location'].map(cities)
rentals
```

```python
rentals['Town/City'].unique()
```

```python
#England, Scotland and Wales are all under the umbrella of the United Kingdom and were wrongly listed as cities
rentals['Town/City'] = rentals['Town/City'].replace([None,'England','Scotland','Wales'], 'Missing')
rentals
```

```python
list(rentals)
```

```python
#Rearranging the order of the columns so the country and city columns are next to the guest location
column_names = ['Listing Name',
 'Guest Name',
 'Guest Location',
 'State/Country',
 'Town/City',               
 'Check-in',
 'Check-Out',
 '# of Adults',
 '# of Children',
 '# of Nights',
 '# of Infants',
 'Currency',
 'Payout',
 'Cleaning Fee',
 'Accommodation Fee',
 'Other Fees',
 'Platform Fee',
 'Reservation Code',
 'Source',
 'Confirmation Date']
rentals = rentals.reindex(columns=column_names)
rentals
```

```python
rentals['Total Revenue'] = rentals['Payout'] + rentals['Platform Fee']
rentals['Cumulative Frequency'] = rentals['Total Revenue'].cumsum()
rentals
```

```python
rentals['Check-in'] = rentals['Check-in'].astype(str)
rentals['Check-Out'] = rentals['Check-Out'].astype(str)
rentals.dtypes
```

```python
# Defining month and year functions that split the dates and return the month and year from them
def month(x):
    x = x.split('-')[1]
    return x

def year(x):
    x = x.split('-')[0]
    return x

#Mapping the functions and creating new columns for year and month
rentals['Year'] = rentals['Check-in'].map(year)
rentals['Month'] = rentals['Check-in'].map(month)
rentals
```

```python
rentals['Year'] = rentals['Year'].astype(int)
rentals['Total Revenue'] = rentals['Total Revenue'].astype(float)
rentals['Cumulative Frequency'] = rentals['Cumulative Frequency'].astype(float)

#Plotting the line graph of cumulative total revenue over time
plt.figure(figsize=(14,6))
plt.title('Cumulative Total Revenue Over Time')
sns.lineplot(x=rentals['Check-in'], y=rentals['Cumulative Frequency'])
plt.xticks(np.arange(10, 350, step=30))
plt.xlabel('Date')
plt.ylabel('Cumulative Total Revenue (£)')
#The graph shows that cumulative revenue has been steadily increasing over the past few years
```

```python
#Plotting bar chart of cumulative total revenue per year
plt.figure(figsize=(14,6))
plt.title('Cumulative Total Revenue Per Year')
sns.barplot(x=rentals['Year'], y=rentals['Cumulative Frequency'])
plt.xlabel('Year')
plt.ylabel('Cumulative Total Revenue (£)')
```

```python
#Plotting the distribution curve of rentals by year
plt.figure(figsize=(14,6))
plt.title('Distribution of Rentals Over Time')
sns.kdeplot(data=rentals['Year'], shade=True)
plt.xlabel('Year')
#The graph shows that a large majority of the rentals were made in 2019
```

```python
#Plotting distribution curve of revenue
plt.figure(figsize=(14,6))
plt.title('Distribution of Revenue')
sns.kdeplot(data=rentals['Total Revenue'], shade=True)
plt.xlabel('Revenue (£)')
#The graph shows a large majority of rentals are between £0 and £500
```

```python
#Creating a 2D kde plot of the two previous distributions
sns.jointplot(x=rentals['Year'], y=rentals['Total Revenue'], kind='kde')
```

```python
#Creating a pivot table of total revenue by check-in and country
geo_revenue = rentals.pivot_table('Total Revenue', index='Check-in', columns='State/Country', aggfunc=sum)
#Filling in missing values with '0'
geo_revenue = geo_revenue.fillna('0')
geo_revenue = geo_revenue.astype(float)
geo_revenue
```

```python
#Checking the sum of all the values in this table matches the final cumulative frequency
#value in the previous table
geo_revenue.sum().sum()
```

```python
#Creating a new column that sums all column values except for United Kingdom
column_list = list(geo_revenue)
column_list.remove('United Kingdom')
column_list
geo_revenue['Rest of World'] = geo_revenue[column_list].sum(axis=1)
geo_revenue
```

```python
#Obtaining the cumulative revenues from the uk and the rest of the world and showing them in a new dataframe
geo_revenue['United Kingdom'] = geo_revenue['United Kingdom'].cumsum()
geo_revenue['Rest of World'] = geo_revenue['Rest of World'].cumsum()
uk_vs_world = geo_revenue[['United Kingdom','Rest of World']]
uk_vs_world
```

```python
#PLotting the line graph of cumulative revenue over time for both uk and rest of world
plt.figure(figsize=(14,6))
plt.title('Cumulative Total Revenue Over Time (UK vs Rest of World)')
sns.lineplot(data=uk_vs_world)
plt.xticks(np.arange(10, 350, step=30))
plt.xlabel('Date')
plt.ylabel('Cumulative Total Revenue (£)')
#The graph shows that total revenue from abroad is increasing at a higher rate than total revenue back in the uk
```

```python
#Creating a pivot table containing each listing name by total payout
payout_pie = rentals.pivot_table('Payout', index='Listing Name', aggfunc=sum)
payout_pie
```

```python
#Plotting the pie chart from the previous table
payout_pie.plot.pie(y='Payout', figsize=(5, 5))
plt.title('Pie chart of Payout by Listing Name (1,2,3,4)')
#The graph shows that the majority of rentals have 2 or 3 floors
```
